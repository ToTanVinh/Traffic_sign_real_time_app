### Split dataset to validation set, training set, test set.

```python
import os
import shutil
import random

# Paths
dataset_path = 'data/vnts'
train_path = os.path.join(dataset_path, 'train')
val_path = os.path.join(dataset_path, 'val')
test_path = os.path.join(dataset_path, 'test')

# Create directories
os.makedirs(os.path.join(train_path, 'images'), exist_ok=True)
os.makedirs(os.path.join(train_path, 'labels'), exist_ok=True)
os.makedirs(os.path.join(val_path, 'images'), exist_ok=True)

os.makedirs(os.path.join(val_path, 'labels'), exist_ok=True)
os.makedirs(os.path.join(test_path, 'images'), exist_ok=True)
os.makedirs(os.path.join(test_path, 'labels'), exist_ok=True)

# Get list of image files
images_path = os.path.join(dataset_path, 'images')
labels_path = os.path.join(dataset_path, 'labels')
all_images = [f for f in os.listdir(images_path) if f.endswith('.jpg')]

# Shuffle the image list
random.shuffle(all_images)

# Split dataset
total = len(all_images)
train_end = int(total * 0.7)
val_end = int(total * 0.85)

train_images = all_images[:train_end]
val_images = all_images[train_end:val_end]
test_images = all_images[val_end:]
```

```python

def move_files(file_list, src_dir_image, src_dir_label, dst_dir_image, dst_dir_label):
    for file_name in file_list:
        # Move image files
        shutil.copy(os.path.join(src_dir_image, file_name), os.path.join(dst_dir_image, file_name))
        # Move label files
        label_file = file_name.replace('.jpg', '.txt')
        shutil.copy(os.path.join(src_dir_label, label_file), os.path.join(dst_dir_label, label_file))

move_files(train_images, images_path, labels_path, os.path.join(train_path, 'images'), os.path.join(train_path, 'labels'))
move_files(val_images, images_path, labels_path, os.path.join(val_path, 'images'),  os.path.join(val_path, 'labels'))
move_files(test_images, images_path, labels_path, os.path.join(test_path, 'images'), os.path.join(test_path, 'labels'))

print("Dataset splitting complete.")
```

```python


# Paths to your dataset (update these to match your actual dataset structure)
dataset_path = 'data/vnts/'
train_images_path = os.path.join(dataset_path, 'train')
val_images_path = os.path.join(dataset_path, 'val')
test_images_path = os.path.join(dataset_path, 'test')  # Optional

# Read classes.txt to get class names
classes_file = os.path.join(dataset_path, 'classes.txt')

with open(classes_file, 'r') as f:
    class_names = f.read().splitlines()

# Number of classes
nc = len(class_names)

# Create dataset.yaml content
dataset_yaml_content = f"""
train: {train_images_path}
val: {val_images_path}
test: {test_images_path}  # Optional

nc: {nc}

names:
"""

for class_name in class_names:
    dataset_yaml_content += f"  - {class_name}\n"

# Write to dataset.yaml file
with open(os.path.join(dataset_path, 'dataset.yaml'), 'w') as f:
    f.write(dataset_yaml_content)

print("dataset.yaml file created successfully.")
```

```python
import zipfile
import os
from IPython.display import FileLink

def zip_dir(directory = os.curdir, file_name = 'directory.zip'):
    """
    zip all the files in a directory
    
    Parameters
    _____
    directory: str
        directory needs to be zipped, defualt is current working directory
        
    file_name: str
        the name of the zipped file (including .zip), default is 'directory.zip'
        
    Returns
    _____
    Creates a hyperlink, which can be used to download the zip file)
    """
    os.chdir(directory)
    zip_ref = zipfile.ZipFile(file_name, mode='w')
    for folder, _, files in os.walk(directory):
        for file in files:
            if file_name in file:
                pass
            else:
                zip_ref.write(os.path.join(folder, file))

    return FileLink(file_name)
zip_dir()
```
